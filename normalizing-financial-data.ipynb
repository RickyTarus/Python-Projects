### **Advanced Financial Data Gathering**

```python
# Major Indices
major_indices = pd.read_html('https://finance.yahoo.com/world-indices/')[0]
major_indices.head()
```

```python
# Grab and create tickers
ticker_list = major_indices['Symbol'].str.replace("^","").str.lower().to_list()
```

```python
df = yf.download(ticker_list, period="1d", start='2015-1-1', end='2018-12-31')
```

```python
df.head()
```

```python
df.columns
```

```python
adj_close = df.dropna(thresh=10, axis=1)['Adj Close']
adj_close.head()
```

```python
adj_close.describe().T
```

```python
adj_close.plot(figsize=(12,12), subplots=True)
```

```python
# Drop MXX due to lack of viable data
adj_close= adj_close.drop('MXX', axis=1)
```

```python
# Do subplots as the data is of different scales
adj_close.plot(figsize=(12,20), subplots=True)
```

```python
# Select one Index and get Quarterly means
dji = adj_close['DJI']
dji.resample('3M').mean()
```

```python
dji
```

```python
# Calculate the daily percent change
dji.shift(1)
dji_per_change = dji / (dji.shift(1) - 1)
dji_per_change.plot(figsize=(12,8))
```

- This can help in checking the volatility of the market

```python
# Calculate the daily rate of return
dji_log_return_shift = np.log(dji / dji.shift(1))
dji_log_return_shift.plot(figsize=(12,8))
```

- Slightly less volatility when you use the rate of return
