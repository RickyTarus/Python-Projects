# Adding the total GAD column of each client into the dataset

```python
df['Total GAD']=df[['GAD1','GAD2','GAD3','GAD4','GAD5','GAD6','GAD7']].sum(axis=1)
df['Mean GAD']= (df['Total GAD'] / 7).round(2)
df.head()
```

# 4.a) Overall mean scores on the GAD.

```python
ovr_mean=df['Mean GAD'].mean().round(2)
print(f'The Overall Mean Score on the GAD is {ovr_mean}')
```

# 4.b)Mean GAD score per tribe ,gender,school and age.

```python
mgs_tribe=pd.DataFrame(df.groupby("Tribe")['Mean GAD'].mean().round(2).reset_index())
mgs_gender=pd.DataFrame(df.groupby('Gender')['Mean GAD'].mean().round(2).reset_index())
mgs_school=pd.DataFrame(df.groupby("School")['Mean GAD'].mean().round(2).reset_index())
mgs_age=pd.DataFrame(df.groupby('Age')['Mean GAD'].mean().round(2).reset_index())
```

```python
print(f'Mean GAD score by tribe is:\n {mgs_tribe}')
```

```python
print(f'Mean GAD score by gender is:\n {mgs_gender}')
```

```python
print(f'Mean GAD school by tribe is:\n {mgs_school}')
```

```python
print(f'Mean GAD score by age is:\n {mgs_age}')
```

# 6.)Visualising relationship Finding between the GAD mean scores and Demographics

# a)Tribe

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 2, 1)
plt.title("Bar Graph showing Tribe and Corresponding GAD mean")
sns.barplot(x='Tribe', y='Mean GAD', data=df,ci=None)

plt.xlabel('Tribe', fontweight='bold', fontsize='11', fontfamily='sans-serif')
plt.ylabel('Mean GAD', fontweight='bold', fontsize='11', fontfamily='sans-serif')
plt.grid(axis='x', alpha=0)
plt.grid(axis='y', alpha=0.4);
```

```python
plt.pie(x=mgs_tribe['Mean GAD'],labels=mgs_tribe['Tribe'],explode=[0.01,0.01],autopct='%1.1f%%')
plt.title("Pie chart showing percentage of Mean GAD by Tribe")
```
